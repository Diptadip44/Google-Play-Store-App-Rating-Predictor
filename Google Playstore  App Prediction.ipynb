```python
data.Installs = data.Installs.apply(lambda x: x.replace(',',''))
data.Installs = data.Installs.apply(lambda x: x.replace('+',''))
data.Installs = data.Installs.apply(lambda x: int(x))
```

```python
data['Installs'].nunique()
```

```python
data['Installs'].head()
```

```python
data.head()
```

```python
data['Reviews'] = data['Reviews'].apply(lambda x: int(x))                        #converting object to int
```

```python
final_review=data[data.Reviews > 5000000]
final_review
```

```python
data1=final_review.drop(columns=['Category','Size','Type','Price','Content Rating','Current Ver','Genres','Last Updated','Android Ver'])
data1.head()
```

```python
user_review['Sentiment'].replace('Positive',1,inplace=True)
user_review['Sentiment'].replace('Neutral',0,inplace=True)
user_review['Sentiment'].replace('Negative',-1,inplace=True)
```

```python
user_review.drop(columns=['Translated_Review','Sentiment_Polarity','Sentiment_Subjectivity'],inplace=True)
```

```python
user_review.head()
```

```python
mask1=user_review[user_review['Sentiment']==1.0]
mask1=mask1['App'].value_counts().reset_index()
mask1=mask1.rename(columns={'index':'App','App':'Sentiment(1)'})
```

```python
mask2=user_review[user_review['Sentiment']==0.0]
mask2=mask2['App'].value_counts().reset_index()
mask2=mask2.rename(columns={'index':'App','App':'Sentiment(0)'})
```

```python
mask3=user_review[user_review['Sentiment']==-1.0]
mask3=mask3['App'].value_counts().reset_index()
mask3=mask3.rename(columns={'index':'App','App':'Sentiment(-1)'})
```

```python
mask=mask2.merge(mask1,left_on='App',right_on='App')
mask=mask3.merge(mask,left_on='App',right_on='App')
mask.head()
```

```python
data2=data1.merge(mask,left_on='App',right_on='App')
data2.head()
```

```python
X=data2.iloc[:,2:7].values
X
```

```python
y=data2.iloc[:,1:2].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=9)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```
